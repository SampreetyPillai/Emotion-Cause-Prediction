```python
import pandas as pd
data = pd.read_csv("annotated_output3.csv")
```

```python
nodes = []

for index, row in data.iterrows():
    nodes.append({
        'id': f"{row['conv_id']}_{row['turn']}_{row['clause_number']}",
        'clause': row['clause'],
        'speaker': row['speaker'],
        'emotion_type': row['annotation'],
        'detected_emotion': row['emotion']
    })
```

```python
missing_clause_nodes = [node for node in nodes if not node.get('clause')]
print(f"Total nodes: {len(nodes)}")
print(f"Nodes missing 'clause': {len(missing_clause_nodes)}")
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
for node in nodes:
    clause = node.get('clause')
    if clause:  
        node['embedding'] = model.encode(clause)
    else:
        print(f"Missing clause for Node ID: {node['id']}")
```

```python
import pandas as pd
nodes_df = pd.DataFrame([
    {
        'id': node['id'],
        'clause': node['clause'],
        'speaker': node['speaker'],
        'emotion_type': node['emotion_type'],
        'detected_emotion': node['detected_emotion'],
        'embedding': ','.join(map(str, node['embedding']))  # Convert array to string
    }
    for node in nodes
])

nodes_df.to_csv('node_embeddings3.csv', index=False)
print("Embeddings saved to 'node_embeddings3.csv'")
```
